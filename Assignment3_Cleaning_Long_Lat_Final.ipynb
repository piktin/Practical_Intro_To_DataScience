```python
import os
import fnmatch
import pandas as pd
import glob 
import re
import numpy as np

```

```python
# define a function to read all stations and make a list of the stations in the file stations.txt
def find_stations():
    f = open('stations.txt', 'r')
    fread = f.readlines()
    stations_list = []
    for station in fread:
        stations_list.append(station.replace('\n',''))
    return stations_list

weather_stations = find_stations()
weather_stations
```

```python
#make a dataframe of all the stations names
pd_station = pd.DataFrame(weather_stations)
pd_station['Stations'] = pd_station
del pd_station[0]
pd_station
```

```python
#create a list of all the stations_names.txt files
mystations = glob.glob('*.txt')
print(mystations)
```

```python
#create a function to take only the first 3 lines of the text files
def read_3line(file):
    with open(file, 'rt') as fd:
        third_line = fd.readlines()[0:3]
    return third_line
 
station_lines = sorted(map(read_3line, mystations))  # apply read first 3 lines function all text files
station_lines
```

```python
lat_ = [float(items) for items in lat_]
lat_
```

```python
#create a list of locations based on latitude and longtitude
location = []
for latitude in lat_:
    if 57.53 <= latitude <= 60.9: 
        loca_name = 'North'
        location.append(loca_name)
    elif 53.56 <= latitude <= 57.23:
        loca_name = 'Central'
        location.append(loca_name)
    elif 49.9 <= latitude <= 53.56:
        loca_name = 'South'
        location.append(loca_name)
location
```
