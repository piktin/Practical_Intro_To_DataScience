```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import io
```

```python
col_names = ('Year','Month','Tmax','Tmin','AF','Rain','Sun', 'Notes')
comment_lines = 5
header = 2
```

```python
aber = pd.read_csv('aberporth.txt',
   skiprows=comment_lines + header,
   header=None,
   names=col_names,
   delimiter=' ',
   skipinitialspace= True,
   error_bad_lines= False)

aber['Sun']=aber['Sun'].str.replace('#','')
aber['AF']=pd.to_numeric(aber['AF'], errors='coerce')
aber['Sun']=pd.to_numeric(aber['Sun'], errors='coerce')
aber['Tmax']=pd.to_numeric(aber['Tmax'], errors='coerce')
aber['Tmin']=pd.to_numeric(aber['Tmin'], errors='coerce')


#Creating Df displaying mean values of all columns, grouped by years
aber_mean = aber.groupby(aber['Year']).mean()
#Slice it only into range of interest: I'm interested in weather data of the last 20 years exluding year 2021
#due to incomplete data
aber_range = aber_mean.loc[2000:2020]

#I add the last row with the means of all the years for each columns with the station name as the row name
aberr = aber_range.append(aber_range.mean().rename('Aberporth'))

#I create a Tmean column, with average between Tmax n Tmin of each year
aberr['Tmean'] = aberr.loc[: , "Tmax":"Tmin"].mean(axis=1)

#I am only interested in the mean row to use for my final df of all stations
aberporth = aberr['Aberporth':]
aberporth
```

```python
def find_stations():
    f = open('stations.txt', 'r')
    fread = f.readlines()
    stations_list = []
    for station in fread:
        stations_list.append(station.replace('\n',''))
    return stations_list

weather_stations = find_stations()
weather_stations

```
