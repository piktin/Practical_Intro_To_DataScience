```python
def lat_band(row):
    if 54.6 < row['Latitude'] <= 55.1 and -2.2 <= row['Longtitude']<= -1.4:
        val = 'NORTH EAST'
    elif 53.3 < row['Latitude'] <= 54.7 and -2.9 <= row['Longtitude'] <= -2.1:
        val = 'NORTH WEST'
    elif 53.2 < row['Latitude'] <= 54.5 and -1.8 <= row['Longtitude'] <= -0.6:
        val = 'YORKSHIRE AND THE HUMBER'
    elif 52.4 < row['Latitude'] <= 53.2 and -1.3 <= row['Longtitude'] <= -0.4:
        val = 'EAST MIDLANDS'
    elif 51.8 <= row['Latitude'] <= 52.8 and -2.8 <= row['Longtitude'] <= -2.3:
        val = 'WEST MIDLANDS'
    elif 52.0 < row['Latitude'] <= 52.6 and -1.2 <= row['Longtitude'] <= 1.8:
        val = 'EAST'
    elif 51.3 <= row['Latitude'] <= 51.6 and -0.5 <= row['Longtitude'] <= -0.1:
        val ='LONDON'
    elif 50.5 < row['Latitude'] <= 51.8 and -1.5 <= row['Longtitude'] <= 1.4 :
        val = 'SOUTH EAST'
    elif 50 < row['Latitude'] <= 51.2:
        val = 'SOUTH WEST'
    elif 51.0 < row['Latitude'] <= 53.3 and -4.7 <= row['Longtitude']<= -3.0:
        val = 'WALES'
    elif 55.2 < row['Latitude']:
        val = 'SCOTLAND'
    elif 54.3 < row['Latitude'] and -6.7 <= row['Longtitude']:
        val = 'NORTHERN IRELAND'
    else:
        val = 'NA'
    return val
```

```python
weather = pd.read_csv('final_weather_Stations.csv')
weather.drop(weather.columns[[0]], axis=1, inplace=True)
weather
```

```python
weather['Area'] = weather.apply(lat_band, axis=1)
weather
```

```python
weather.to_csv('weather.csv')
```
