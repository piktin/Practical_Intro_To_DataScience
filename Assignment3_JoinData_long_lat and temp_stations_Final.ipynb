```python
import pandas as pd
```

```python
final_weather_stations = pd.read_csv('final_weather_stations.csv')
del final_weather_stations['Unnamed: 0']
final_weather_stations
```

```python
final_stations = pd.read_csv('final_stations.csv')
del final_stations['Unnamed: 0']
final_stations
```

```python
# join the 2 dataframes of weather and stations, removing Notes and Year as these columns have mainly NANs values
uk_stations = pd.merge(final_stations, final_weather_stations, left_index=False,right_index=False, how='left')
del uk_stations['Notes']
del uk_stations['Year']
uk_stations
```

```python
uk_stations.to_csv('uk_stations.csv')
```
